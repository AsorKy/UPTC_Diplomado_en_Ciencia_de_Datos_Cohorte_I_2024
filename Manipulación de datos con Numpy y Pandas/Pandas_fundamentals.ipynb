Pandas proporciona varias funciones para cargar datos desde diferentes fuentes, como archivos CSV, Excel, bases de datos SQL, JSON, HTML, etc. Algunas de las funciones más comunes para cargar datos incluyen:

### 1.3. Manejo de datos nulos


---
Manejo de datos nulos

---

Los datos nulos, también conocidos como valores faltantes o `NaN` (Not a Number), son valores que faltan o no están presentes en un conjunto de datos. En Pandas, los datos nulos se representan como `NaN`. Es importante tener en cuenta los datos nulos por las siguientes razones:

* De no ser corregidos, pueden comprometer la intergridad de los datos e ingresar sesgos en el análisis de los mísmos.

* De no ser corregidos, comprometen la presición del análisis y del modelo de predicción.

* Los datos nulos pueden tener significados diferentes dependiendo del contexto del conjunto de datos. Al comprender y manejar los datos nulos adecuadamente, se puede interpretar correctamente el conjunto de datos y tomar decisiones informadas basadas en los resultados del análisis.

```python
# Creación de un dataframe con datos nulos
import numpy as np
import pandas as pd

datos = {'var_1':[1,2,3,np.nan],
         'var_2':[4, np.nan,6,7],
         'var_3':['a','b','c', None],
         'var_4':[1,4,3,2]}
df_test = pd.DataFrame(datos)
df_test
```

Para identificar que columnas y el número de valores nulos que posee cada variable, podemos emplear el método `.info()`.

```python
# Detección de valores faltantes mediante .info()
df_test.info()
```

Por otra parte, es posible identificar concretamente la posición de los valores faltanes mediante el método `.isnull()`.

```python
# Identificar datos nulos en el dataframe
df_test.isnull()
```

```python
# Identificar datos nulos con un valor numérico
df_test.isnull()*1
```

```python
# Identificar datos no nulos en un dataset
df_test.notnull()
```

```python
# Determinar la cantidad de valores nulos o faltantes por columna
df_test.isnull().sum()
```

El manejo de valores nulos o faltantes es un tema bastante amplio, que implica procesos de interpretación de dichos valores para determinar su relevancia y procesos relacionados con la imputación de estos valores según diferentes técnicas. De forma general, cuando nos enfrentamos a un dataset con valores nulos o faltantes, pueden tomarse las siguientes acciones según el contexto de los datos y la finalidad de la acción:

* Eliminación de datos faltantes: Posible mediante métodos como `.dropna()`.
* Reemplazo de valores faltantes: Posible mediante métodos como `.fillna().
* Imputación de datos faltantes: Posible mediante métodos como `.fillna()` e `.interpolate()`.

```python
# Eliminación de valores faltantes (eliminación por filas)
df_test_1 = df_test.copy()
df_test_1.dropna(axis = 0, inplace=True)
df_test_1
```

```python
# Eliminación de valores faltantes (eliminación por columnas)
df_test_2 = df_test.copy()
df_test_2.dropna(axis=1, inplace=True)
df_test_2
```

```python
# Remplazo de valores faltantes por un string
df_test_1 = df_test.copy()
df_test_1.fillna('Missing', inplace=True)
df_test_1
```

```python
# Remplazo de valores faltantes por un string
df_test_1 = df_test.copy()
df_test_1['var_1'].fillna(0.0, inplace=True)
df_test_1['var_2'].fillna(0.0, inplace=True)
df_test_1['var_3'].fillna('Missing', inplace=True)
df_test_1
```

```python
df_test_1.info()
```

```python
# Imputación de valores por promedio
df_test_1 = df_test.copy()
df_test_1.fillna(df_test_1.mean(), inplace=True)
df_test_1
```

El método `.interpolate()` en Pandas se utiliza para interpolar valores faltantes en un DataFrame. La interpolación es el proceso de estimar valores desconocidos en función de valores conocidos mediante métodos como regresión lineal o polinomial.

```python
# Imputación de valores por interpolate
df_test_1 = df_test.copy()
df_test_1.interpolate(inplace=True)
df_test_1
```

```python
# Crear un DataFrame con valores faltantes
data = {'A': [1, 2, None, 4, None, 6],
        'B': [None, 5, None, None, 8, 9]}
df_incomplete = pd.DataFrame(data)

print(df_incomplete)
# Interpolar valores faltantes utilizando interpolación lineal
df_interpolado = df_incomplete.interpolate(method='linear', axis=0)
df_interpolado
```

### 1.4. Manejo de duplicados

---
Manejo de duplicados

---

Los valores duplicados son elementos que aparecen más de una vez en un conjunto de datos. Estos duplicados pueden estar presentes en una o más columnas del conjunto de datos y pueden representar una variedad de situaciones, como errores de entrada, registros duplicados o datos repetidos. Detectar y manejar valores duplicados es fundamental en el análisis de datos por varias razones:

* Asegurar la integridad y calidad de los datos: Los valores duplicados pueden distorsionar la integridad de los datos y conducir a análisis incorrectos. Identificar y eliminar estos duplicados ayuda a garantizar que los resultados del análisis sean precisos y confiables.

* Asegurar la consistencia de los resultados: Al eliminar los valores duplicados, se asegura que los resultados del análisis sean coherentes y replicables.

* Asegurar la eficiencia en el procesamiento de datos: Los valores duplicados pueden aumentar la complejidad y el tiempo de procesamiento durante el análisis de datos. Al eliminar los duplicados, se simplifica el conjunto de datos y se mejora la eficiencia en el procesamiento y la manipulación de los datos.

```python
# Creación de un dataset con valores duplicados
data = {
    'A': [1, 2, 3, 4, 5, 2, 7, 8, 9, 2],
    'B': ['a', 'b', 'c', 'd', 'e', 'b', 'g', 'h', 'i', 'b'],
    'C': [10.5, 20.3, 15.8, 7.2, 3.4, 20.3, 18.9, 22.1, 30.6, 20.3]
}

df_test = pd.DataFrame(data)
df_test
```

La detección de valores duplicados se da con base a la existencia de regístros cuyos valores de variables son exactamente los mísmos.

```python
# Detectar filas duplicadas en el DataFrame
duplicados = df_test.duplicated()
duplicados
```

No obstante, antes de proceder, es importante saber que es posible detectar los valores únicos dentro de cada columna.

```python
# Valores únicos dentro de una columna
df_test['B'].unique()
```

```python
# Cuántos valores posibles toma una variable
df_test['B'].nunique()
```

En general, el manejo de duplicados suele consistir en la eliminación de los mísmos. Para ello, empleamos el método `.drop_duplicates()`.

```python
# Eliminación simple de duplicados
df_test_1 = df_test.copy()
df_test_1.drop_duplicates(inplace=True)
df_test_1
```

Para reindexar el dataframe resultante después de eliminar los duplicados, empleamos el parámetro `ignore_index`.

```python
# Eliminación simple de duplicados
df_test_1 = df_test.copy()
df_test_1.drop_duplicates(inplace=True, ignore_index=True)
df_test_1
```

Dentro de Pandas, es posible eliminar duplicados de acuerdo con un conjunto de variables o columnas especificas mediante el parámetro `subset`.

```python
# Eliminación de duplicados de acuerdo con un conjunto de columnas
data = {
    'A': [1, 2, 3, 2, 4, 5, 2],
    'B': ['a','b','c','b','e','f','b'],
    'C': np.random.randn(7)
}

df_test = pd.DataFrame(data)
print(df_test)

df_subset = df_test.drop_duplicates(subset = ['A','B'], ignore_index=True)
df_subset
```

El parámetro `keep` en `drop_duplicates()` permite especificar qué duplicados mantener. Las opciones son `'first'` para mantener la primera aparición de cada duplicado, `'last'` para mantener la última aparición y `False` para eliminar todas los duplicados.

```python
# Uso del comando Keep
df_subset = df_test.drop_duplicates(subset=['A','B'], keep='last')
df_subset
```

### 1.5. Combinación y agrupación de datos

---
Combinación de datos

---

La combinación de DataFrames y la agrupación de datos son dos técnicas esenciales en el análisis y manipulación de datos con Pandas. Estas operaciones permiten realizar tareas complejas de manejo y análisis de datos de manera eficiente y efectiva.

La combinación de DataFrames implica la unión de dos o más conjuntos de datos basados en una o más columnas comunes. Esto permite combinar la información de diferentes fuentes de datos en un solo DataFrame, lo que facilita el análisis y la interpretación de los datos. Las operaciones comunes de combinación de DataFrames incluyen la fusión (`merge`), la concatenación (`concat`) y la unión (`join`).

* Fusión (`merge`): Combina filas de dos DataFrames basándose en una columna común.
* Concatenación (`concat`): Une dos o más DataFrames a lo largo de un eje específico (filas o columnas).
* Unión (`join`): Combina columnas de dos DataFrames basándose en sus índices.

```python
# Crear DataFrames de ejemplo
df1 = pd.DataFrame({'A': [1, 2, 3],
                    'B': ['a', 'b', 'c']})

df2 = pd.DataFrame({'C': [4, 5, 6],
                    'D': ['d', 'e', 'f']})
```

```python
# Concatenación Vertical (a lo largo de las filas)
result_vertical = pd.concat([df1, df2], axis=0, ignore_index=True)
print("Concatenación Vertical:")
result_vertical
```

```python
# Concatenación Horizontal (a lo largo de las columnas)
result_horizontal = pd.concat([df1, df2], axis=1)
print("Concatenación Horizontal:")
result_horizontal
```

Las fusiones de dataframes se consísten en combinar dataframes de acuerdo a características comúnes a los dataframes implicados a diferencia de la concatenación la cuál simplemente consíste en unir dataframes bien sea vertical u horizontalmente.

```python
# DataFrame df1
df1 = pd.DataFrame({'clave': ['A', 'B', 'C', 'D'],
                    'valor': [1, 2, 3, 4]})

# DataFrame df2
df2 = pd.DataFrame({'clave': ['B', 'D', 'E', 'F'],
                    'otro_valor': [5, 6, 7, 8]})
print(df1)
print(df2)
```

En la fusión por columna común, los DataFrames se fusionan utilizando una columna que es común a ambos DataFrames. Esta columna se especifica utilizando el argumento `on`. La fusión combina las filas de los DataFrames donde los valores en la columna común coinciden.

```python
# Fusión por columna común
merged_df = pd.merge(df1, df2, on='clave')
print("Fusión por Columna Común:")
merged_df
```

En la fusión interna, también conocida como fusión por defecto, se conservan solo las filas que tienen claves coincidentes en ambos DataFrames. Es decir, solo se incluyen las filas donde los valores en la columna común coinciden entre los DataFrames. Las filas que no tienen coincidencias en ambos DataFrames se descartan.

```python
# Fusión interna
merged_inner = pd.merge(df1, df2, on='clave', how='inner')
print("Fusión Interna:")
merged_inner
```

En la fusión externa, se conservan todas las filas de ambos DataFrames, incluso aquellas que no tienen coincidencias en la columna común. Se crean valores nulos (`NaN`) en las columnas correspondientes para las filas que no tienen coincidencias en el otro DataFrame.

```python
# Fusión externa
merged_outer = pd.merge(df1, df2, on='clave', how='outer')
print("Fusión Externa:")
merged_outer
```

En la fusión izquierda, también conocida como fusión izquierda externa, se conservan todas las filas del DataFrame izquierdo y solo las filas del DataFrame derecho que tienen claves coincidentes. Las filas que no tienen coincidencias en el DataFrame derecho tendrán valores nulos en las columnas correspondientes.

```python
# Fusión izquierda
merged_left = pd.merge(df1, df2, on='clave', how='left')
print("Fusión Izquierda:")
merged_left
```

En la fusión derecha, también conocida como fusión derecha externa, se conservan todas las filas del DataFrame derecho y solo las filas del DataFrame izquierdo que tienen claves coincidentes. Las filas que no tienen coincidencias en el DataFrame izquierdo tendrán valores nulos en las columnas correspondientes.

```python
# Fusión derecha
merged_right = pd.merge(df1, df2, on='clave', how='right')
print("Fusión Derecha:")
merged_right
```

El método `join()` en Pandas proporciona una forma conveniente de combinar DataFrames utilizando sus índices como clave de unión. Aunque es menos flexible que `merge()`, es útil cuando necesitas fusionar DataFrames basados en sus índices sin tener que especificar columnas específicas. La diferencia principal entre `merge()` y `join()` radica en cómo se selecciona la clave de unión: columnas específicas para `merge()` e índices para `join()`.

```python
# DataFrame df1
df1 = pd.DataFrame({'clave': ['A', 'B', 'C', 'D'],
                    'valor': [1, 2, 3, 4]}
                   , index = ['a','b','c','d']
                   )

# DataFrame df2
df2 = pd.DataFrame({'clave': ['B', 'D', 'E', 'F'],
                    'otro_valor': [5, 6, 7, 8]}
                   , index = ['c','d','e','f']
                   )
print(df1)
print(df2)
```

```python
# Inner Join
try:
  inner_join = df1.join(df2, how='inner')
except:
  print('Agrega los sufijos izquierdo y derecho para diferenciar columnas:')
  inner_join = df1.join(df2, how='inner', lsuffix='_left', rsuffix='_right')

print("Inner Join:")
inner_join
```

```python
# Outer Join
try:
  outer_join = df1.join(df2, how='outer')
except:
  print('Agrega los sufijos izquierdo y derecho para diferenciar columnas:')
  outer_join = df1.join(df2, how='outer', lsuffix='_left', rsuffix='_right')

print("Outer Join:")
outer_join
```

```python
# Left Join
try:
  left_join = df1.join(df2, how='left')
except:
  print('Agrega los sufijos izquierdo y derecho para diferenciar columnas:')
  left_join = df1.join(df2, how='left', lsuffix='_left', rsuffix='_right')

print("left Join:")
left_join
```

```python
# right Join
try:
  right_join = df1.join(df2, how='right')
except:
  print('Agrega los sufijos izquierdo y derecho para diferenciar columnas:')
  right_join = df1.join(df2, how='right', lsuffix='_left', rsuffix='_right')

print("right Join:")
right_join
```

Agrupación de datos

---
La agrupación de datos implica dividir los datos en grupos basados en ciertas características y luego aplicar una función a cada grupo de datos. Esta técnica es útil para resumir y analizar datos de manera eficiente, especialmente cuando se trabaja con conjuntos de datos grandes. La operación principal para la agrupación de datos es el método `groupby()`, que permite dividir un DataFrame en grupos basados en los valores de una o más columnas y luego aplicar funciones de agregación a estos grupos.


```python
# Datos de ejemplo
tips_data = sns.load_dataset('tips')
tips_data.info()
```

```python
# Agrupación por variable categórica
grupos = tips_data.groupby('day')
tips_data.day.cat.categories
```

Para obtener un grupo en específico, empleamos el método `get_group()`.

```python
# Obtenemos los grupos
grupos_lista = []
for nombre_grupo, grupo in grupos:
  grupos_lista.append(grupo)

grupos_lista[0].head()
```

Una vez los grupos con base a una variable son identificados, es posible aplicar métodos a cada grupo.

```python
# Promedios para cada agrupación
grupos.mean()
```

```python
# Estadísticos básicos para cada agrupación
grupos.describe()['total_bill']
```


El comando `.agg()` en Pandas se utiliza para realizar agregaciones de datos en un DataFrame o en grupos de datos de tal manera que permite calcular una o más estadísticas agregadas para una o más columnas, lo que facilita el resumen y análisis de grandes conjuntos de datos.

```python
tips_data.columns
```

```python
# Agregación y estadísticos
grupos = tips_data.groupby('day')
variables_de_interes = {'total_bill': 'mean', 'tip': 'mean'}
sub_grupos = grupos.agg(variables_de_interes)

sub_grupos
```

El comando `.agg` tambien permite generar agrupaciones con múltiples estadísticas.

```python
# Agregación con múltiples estadísticas
grupos = tips_data.groupby('day')
consulta = {'total_bill' : ['min' , 'max'], 'tip': ['mean','median']}
consulta_anidada = grupos.agg(consulta)

consulta_anidada
```

Finalmente, el comando `.groupby` permite realizar agrupaciones anidadas, es decir, agrupaciones con base a múltiples variables.

```python
# Agrupaciones con base a múltiples variables
variables_a_agrupar = ['day', 'time', 'sex']
grupos = tips_data.groupby(variables_a_agrupar)['tip'].mean()

grupos
```

```python
# Agrupaciones con base a múltiples variables
variables_a_agrupar = ['day', 'time', 'sex']
grupos = tips_data.groupby(variables_a_agrupar)['tip'].describe()

grupos
```

Métodos apply

---


El método `apply()` en Pandas se utiliza para aplicar una función arbitraria a los elementos de un DataFrame o una Serie. Esta función puede ser una función integrada de Python, una función definida por el usuario o una función lambda. El método `apply()` es útil para realizar operaciones personalizadas en los datos de un DataFrame y para realizar transformaciones más complejas que no están directamente disponibles mediante los métodos integrados de Pandas. Su sintaxis básica es `DataFrame.apply(func, axis=0, raw=False, result_type=None, args=(), **kwargs)`.

* `func`: Función a aplicar a cada elemento del DataFrame o la Serie.

* `axis`: Eje a lo largo del cual se aplica la función. `0` para aplicar la función a cada columna y `1` para aplicarla a cada fila (solo en DataFrame).

* `raw`: Si es `True`, pasa cada fila o columna como una serie a la función. Por defecto es `False`.

* `result_type`: Tipo de datos de retorno esperado después de aplicar la función. Puede ser `'expand'`, `'reduce'` o `'broadcast'`.

* `args, kwargs`: Argumentos adicionales que se pasan a la función.


```python
# Crear DataFrame de ejemplo
data = {
    'Nombre': ['Juan', 'María', 'Pedro', 'Ana'],
    'Departamento': ['Ventas', 'TI', 'Ventas', 'RRHH'],
    'Salario': [50000, 60000, 55000, 48000]
}

df_test = pd.DataFrame(data)

# Definir una función para aplicar al salario
def aumento_salario(salario):
    return salario * 1.05  # Incremento del 5%

# Aplicar la función al salario usando apply()
df_test['Salario Actualizado'] = df['Salario'].apply(aumento_salario)

# Mostrar el DataFrame con el salario actualizado
df_test
```
